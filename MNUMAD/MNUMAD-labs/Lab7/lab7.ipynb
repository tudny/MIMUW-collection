```python
def read_file(filename: str) -> [str]:
    with open(filename, 'r') as f:
        return f.read().splitlines()
```

```python
docs = ["data/wc.txt", "data/jp.txt", "data/pw.txt"]
docs_text = [read_file(doc) for doc in docs]
```

2. Zamienić listę dokumentów na macierz częstości
    Użyć `from sklearn.feature_extraction.text import TfidfVectorizator`

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import requests

POLISH_STOP_WORDS_URL = "https://raw.githubusercontent.com/bieli/stopwords/master/polish.stopwords.txt"

def download_file(url: str):
    response = requests.get(url)
    return response.content.splitlines()

POLISH_STOP_WORDS = download_file(POLISH_STOP_WORDS_URL)


def vectorize(docs: [str]) -> [str]:
    vectorizer = TfidfVectorizer(
        stop_words=POLISH_STOP_WORDS,
    )
    return vectorizer.fit_transform(docs), vectorizer

docs_vectorized_with_vectorizer = [vectorize(doc) for doc in docs_text]
```

```python
import numpy as np

k = 7
n = 5
```

```python
def process(matrix, vectorizer):
    u, s, vh = np.linalg.svd(matrix.todense())
    u_k = u[:, :k]
    s_k = s[:k]
    vh_k = vh[:k, :]
    dictionary = vectorizer.get_feature_names_out()
    for i in range(k):
        v_i = vh_k[i, :]
        res_sorted = np.argsort(v_i)
        names = dictionary[res_sorted]
        print(names[:n])
```

```python
processed = [process(matrix, vectorizer) for (matrix, vectorizer) in docs_vectorized_with_vectorizer]
```
